```python
import pandas as pd
```

```python
df = pd.read_csv("/home/shweta/rawdata.csv")
```

```python
df
```

```python
from datetime import datetime
import time

timestamps = []
date_format = "%m/%d/%Y %I:%M:%S %p"
for i in range(len(df)):
    date_str = df.iloc[i]['date'] + " " +df.iloc[i]['time']
    dt = datetime.strptime(date_str, date_format)
    epoch_timestamp = int(time.mktime(dt.timetuple()))
    timestamps.append(epoch_timestamp)
```

```python
timestamps
```

```python
df['timestamp'] = timestamps
```

```python
df
```

```python
df = df.sort_values(by=['timestamp'],ascending=[True])
```

```python
output = {
    "1/16/2024": {
        "inside_duration": 0,
        "outside_duration": 0
    },
    "1/17/2024": {
        "inside_duration": 0,
        "outside_duration": 0
    },
    "1/18/2024": {
        "inside_duration": 0,
        "outside_duration": 0
    }
}

previous_timestamp = df.iloc[0]['timestamp']
previous_date = df.iloc[0]['date']
previous_position = df.iloc[0]['position']

for i in range(1, len(df)):
    current_date = df.iloc[i]['date']
    current_position = df.iloc[i]['position']
    current_timestamp = df.iloc[i]['timestamp']
    

    if(current_date == previous_date and previous_position == 'inside' and (current_position == 'inside' or current_position == 'outside')):
        duration = current_timestamp - previous_timestamp
        output[current_date]['inside_duration'] = output[current_date]['inside_duration'] + duration
        
    # condition for outside duration
    if(current_date == previous_date and previous_position == 'outside') and (current_position == 'Inside' or current_position == 'outside'):
        duration = current_timestamp - previous_timestamp
        output[current_date]['outside_duration'] = output[current_date]['outside_duration'] + duration 


    previous_timestamp = current_timestamp
    previous_date = current_date
    previous_position = current_position
    
    
    
        
```

```python
output
```

```python
output2 = {
    "1/16/2024": {
        "pick_activities": 0,
        "place_activities": 0
    },
    "1/17/2024": {
        "pick_activities": 0,
        "place_activities": 0
    },
    "1/18/2024": {
        "pick_activities": 0,
        "place_activities": 0
    }
}

for i in range(len(df)):
    if df.iloc[i]['date'] == "1/16/2024":
        if df.iloc[i]['activity'] == "picked":
            output2['1/16/2024']['pick_activities'] += 1
        else:
            output2['1/16/2024']['place_activities'] += 1

    if df.iloc[i]['date'] == "1/17/2024":
        if df.iloc[i]['activity'] == "picked":
            output2['1/17/2024']['pick_activities'] += 1
        else:
            output2['1/17/2024']['place_activities'] += 1

    if df.iloc[i]['date'] == "1/18/2024":
        if df.iloc[i]['activity'] == "picked":
            output2['1/18/2024']['pick_activities'] += 1
        else:
            output2['1/18/2024']['place_activities'] += 1
```

```python
output2
```

```python
data = [["1/16/2024",output2['1/16/2024']['pick_activities'],output2['1/16/2024']['place_activities'],output['1/16/2024']['inside_duration'],output['1/16/2024']['outside_duration']],
        ["1/17/2024",output2['1/17/2024']['pick_activities'],output2['1/17/2024']['place_activities'],output['1/17/2024']['inside_duration'],output['1/17/2024']['outside_duration']],
        ["1/18/2024",output2['1/18/2024']['pick_activities'],output2['1/18/2024']['place_activities'],output['1/18/2024']['inside_duration'],output['1/18/2024']['outside_duration']]
        ]
```

```python
df1 = pd.DataFrame(data, columns = ["date","pick_activities","place_activities","inside_duration","outside_duration"])
```
